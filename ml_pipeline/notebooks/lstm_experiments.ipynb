# LSTM Experiments

Load and visualize LSTM training history, loss curves, and accuracy.

```python
import json
import matplotlib.pyplot as plt
from pathlib import Path

eval_path = Path('../models/saved_models/lstm_evaluation.json')

if not eval_path.exists():
    print('Run train_lstm.py first to generate evaluation data.')
else:
    with open(eval_path) as f:
        results = json.load(f)

    print(f'Model      : {results["model"]}')
    print(f'Accuracy   : {results["accuracy"]:.4f}')
    print(f'F1 Score   : {results["f1_score"]:.4f}')
    print(f'Best Epoch : {results["best_epoch"]}')
```

```python
history = results['history']
epochs  = list(range(1, len(history['train_loss']) + 1))

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Loss curves
axes[0].plot(epochs, history['train_loss'], 'b-o', ms=3, label='Train Loss')
axes[0].plot(epochs, history['val_loss'],   'r-o', ms=3, label='Val Loss')
axes[0].axvline(results['best_epoch'], color='g', linestyle='--', alpha=0.7,
                label=f'Best Epoch ({results["best_epoch"]})')
axes[0].set_title('Training & Validation Loss', fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Loss')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Accuracy
axes[1].plot(epochs, history['val_acc'], 'g-o', ms=3, label='Val Accuracy')
axes[1].set_title('Validation Accuracy', fontweight='bold')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Accuracy')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('../models/saved_models/lstm_training_curves.png', dpi=120)
plt.show()
```
