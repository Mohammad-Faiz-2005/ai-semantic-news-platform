# SBERT Experiments

Visualize SBERT embeddings with PCA and t-SNE, and explore similarity examples.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_20newsgroups
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sentence_transformers import SentenceTransformer

CATEGORIES = ['rec.sport.hockey', 'sci.space', 'talk.politics.misc', 'sci.med']
data = fetch_20newsgroups(subset='test', categories=CATEGORIES,
                          remove=('headers','footers','quotes'), random_state=42)

# Sample 200 per category for speed
from collections import defaultdict
samples = defaultdict(list)
for text, label in zip(data.data, data.target):
    if len(samples[label]) < 200:
        samples[label].append((text, label))

texts, labels = zip(*[item for items in samples.values() for item in items])

print(f'Total samples: {len(texts)}')
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(list(texts), normalize_embeddings=True, show_progress_bar=True)
```

```python
# PCA visualization
pca = PCA(n_components=2, random_state=42)
emb_2d = pca.fit_transform(embeddings)

plt.figure(figsize=(10, 7))
colors = ['#3b82f6', '#ef4444', '#10b981', '#f59e0b']

for i, cat in enumerate(CATEGORIES):
    mask = np.array(labels) == i
    plt.scatter(emb_2d[mask, 0], emb_2d[mask, 1],
                c=colors[i], label=cat.split('.')[-1], alpha=0.6, s=15)

plt.title('SBERT Embeddings — PCA 2D Projection', fontsize=14, fontweight='bold')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('../models/saved_models/pca_embeddings.png', dpi=120)
plt.show()
```
