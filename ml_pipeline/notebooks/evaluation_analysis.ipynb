# Evaluation Analysis

Load full_evaluation.json and produce cross-model comparison charts.

```python
import json
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

eval_path = Path('../models/saved_models/full_evaluation.json')

if not eval_path.exists():
    print('Run evaluate_models.py first.')
else:
    with open(eval_path) as f:
        data = json.load(f)

    models = data['models']
    print(f'Loaded {len(models)} model results.')
    for m in models:
        print(f'  {m["model_name"]}: accuracy={m["accuracy"]}, f1={m["f1_score"]}')
```

```python
names   = [m['model_name'] for m in models]
metrics = ['accuracy', 'precision', 'recall', 'f1_score']
colors  = ['#f59e0b', '#3b82f6']
x       = np.arange(len(metrics))
width   = 0.35

fig, ax = plt.subplots(figsize=(10, 5))

for i, (model, color) in enumerate(zip(models, colors)):
    values = [model[m] for m in metrics]
    bars = ax.bar(x + i * width, values, width,
                  label=model['model_name'],
                  color=color, alpha=0.85,
                  edgecolor='white', linewidth=0.5)

    for bar, val in zip(bars, values):
        ax.text(bar.get_x() + bar.get_width()/2,
                bar.get_height() + 0.005,
                f'{val:.2f}',
                ha='center', va='bottom',
                fontsize=8, fontweight='bold')

ax.set_xticks(x + width / 2)
ax.set_xticklabels([m.replace('_', ' ').title() for m in metrics])
ax.set_ylim(0, 1.1)
ax.set_ylabel('Score')
ax.set_title('TF-IDF vs SBERT — Classification Metrics',
             fontsize=13, fontweight='bold')
ax.legend()
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('../models/saved_models/model_comparison.png', dpi=120)
plt.show()
```
